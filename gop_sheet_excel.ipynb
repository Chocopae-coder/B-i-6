```python
import pandas as pd

file_path = "251208_psc.xlsx"
xls = pd.ExcelFile(file_path)

df_all = []
for sheet in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet)
    df["sheet_name"] = sheet  # optional
    df_all.append(df)

result = pd.concat(df_all, ignore_index=True)
result.to_excel("251208_psc_file_tong.xlsx", index=False)
```
